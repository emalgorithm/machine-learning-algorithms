```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class GaussianProcess():
    def __init__(self, mean_function=lambda x: np.zeros((len(x),)), kernel=lambda x, y: np.exp(- (np.linalg.norm(x-y) ** 2) / 2)):
        self.mean_function = mean_function
        self.kernel = kernel
    def sample(self, x, size):
        mu = self.mean_function(x)
        sigma = np.zeros((len(x), len(x)))
        for i in range(len(x)):
            for j in range(len(x)):
                sigma[i, j] = self.kernel(x[i], x[j])

        y = np.random.multivariate_normal(mu, sigma, size)
        return y.T
```

```python
gp = GaussianProcess()
```

```python
x = [i for i in range(100)]
y = gp.sample(x, size=5)
```

```python
plt.plot(x, y)
plt.show()
```
